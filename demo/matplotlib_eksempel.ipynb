```python
import matplotlib.pyplot as plt
```

<p>Vil lage et plott av funksjonen</p>
<blockquote><p>$f(x) = x² $</p>
<p>for området $[-10, 10]$</p>
</blockquote>

<p>Metode:</p>
<ul>
<li>$x$-verdiene legges i listen $X$</li>
<li>$f(x)$-verdiene legges i listen $Y$</li>
</ul>

```python
X = list(range(-10, 11))
print(X)
```

```python
Y = []

for x in X:
    Y.append(x**2)

print(Y)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.plot(X, Y)
plt.axis([-11, 11, -10, 110])
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.plot(X, Y, 'o')       # o, ., x, X, etc
plt.axis([-11, 11, -10, 110])
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
